```python
def extract_text_from_pdf(pdf_path):
    bold_texts = []
    normal_texts = []

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            for char in page.chars:
                if 'Bold' in char['fontname']:
                    bold_texts.append(char['text'])
                else:
                    normal_texts.append(char['text'])

    return bold_texts, normal_texts

pdf_path = '/content/pdf_minery/diccionario_suenios.pdf'
bold_texts, normal_texts = extract_text_from_pdf(pdf_path)

print("Bold Texts:", bold_texts)
print("Normal Texts:", normal_texts)
```

```python
def extract_words_and_meanings(pdf_path):
    bold_words = []
    normal_texts = []

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            # Use extract_words with extra_attrs=['fontname'] to include fontname in the output
            words = page.extract_words(extra_attrs=['fontname'])
            for word in words:
                # Check if 'fontname' key exists before accessing it
                if 'fontname' in word and 'Bold' in word['fontname']:
                    bold_words.append(word['text'])
                else:
                    normal_texts.append(word['text'])

    return bold_words, normal_texts

# Replace '/pdf_minery/diccionario_suenios.pdf' with the path to your PDF file
pdf_path = '/content/pdf_minery/diccionario_suenios.pdf'
bold_words, normal_texts = extract_words_and_meanings(pdf_path)

print("Bold Words:", bold_words)
print("Normal Texts:", normal_texts)
```

#### Removing numbers with regular expressions

To clean the numbers from the "meanings" Series in your DataFrame without deleting the rest of the content, you can use regular expressions to remove any numeric characters. Here's how you can do it in your Google Colab notebook:

```python
import re

# Sample DataFrame
data = {'meanings': ['Dream about 123', '456 Symbolism', 'Just text', '789']}
df_test = pd.DataFrame(data)

# Function to remove numbers from a string
def remove_numbers(text):
    return re.sub(r'\d+', '', text)

# Apply the function to the 'meanings' Series
df_test['meanings'] = df_test['meanings'].apply(remove_numbers)

# Display the cleaned DataFrame
print(df_test)
```
